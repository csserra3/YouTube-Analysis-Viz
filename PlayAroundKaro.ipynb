```python
import altair as alt
import pandas as pd
import os
```

```python
data = pd.read_csv("Data/Global YouTube Statistics.csv", encoding='latin-1')
data = data.dropna()
data.head()
```

```python
selection = alt.selection_point(fields=["channel_type"])


scatterplot = alt.Chart(data).mark_circle().encode(
    x = 'subscribers:Q',
    y = 'video views:Q',
    size = "highest_yearly_earnings:Q",
    color = alt.condition(selection, "channel_type", alt.value('lightgray')),
    tooltip =  alt.Tooltip('Youtuber')
).add_params(selection).properties(width=300, height=300)

# Create a histogram
hist = alt.Chart(data).mark_bar().encode(
    x = alt.X("subscribers", bin=alt.Bin(maxbins=30)),
    y = "count()"
).transform_filter(selection).properties(width=300, height=300)

# Connect our charts using the pipe operation
scatterplot | hist
```

```python
top10_sub = data.sort_values('subscribers', ascending=False).head(10)
top10_sub 
```

```python
alt.Chart(top10_sub).mark_bar().encode(
    x = alt.X("Youtuber").sort('-y'),
    y = alt.Y("subscribers"),
    color = "channel_type"
).properties(width=300, height=300)
```

```python
selection = alt.selection_point(fields=["channel_type"])


scatterplot = alt.Chart(data).mark_circle().encode(
    x = alt.X('subscribers:Q', title = 'Number of subscribers'),
    y = alt.Y('average(highest_yearly_earnings):Q', title = 'Highest Average Yearly Earnings'),
    size = alt.Size("video views:Q", title = 'Number of video views'),
    color = alt.condition(selection, "channel_type", alt.value('lightgray')),
    tooltip =  alt.Tooltip(['Youtuber','subscribers','video views','highest_yearly_earnings'])
).add_params(selection).properties(width=300, height=300, title = 'The Highest Average Yearly Earnings of YouTube Channels by Subscribers')


hist = alt.Chart(data).mark_bar().encode(
    x = alt.X("Youtuber").sort('-y'),
    y = alt.Y("subscribers", title = 'Number of subscribers'),
    color = alt.Color("channel_type", title = 'Type of Channel'),
    tooltip =  alt.Tooltip(['Youtuber','subscribers','video views','highest_yearly_earnings'])
).transform_filter(selection).transform_window(
    rank='rank(subscribers)',
).transform_filter(
    alt.datum.rank <= 10
).properties(width=300, height=300, title = 'Top 10 YouTube Channels based on Subscribers')



scatterplot | hist
```

```python
dropdown = alt.binding_select(
    options=['uploads','highest_yearly_earnings', 'lowest_yearly_earnings', 'subscribers'],
    name='Heatmap color attribute '
)
color_param = alt.param(
    value='highest_yearly_earnings',
    bind=dropdown
)

alt.Chart(data).mark_rect().encode(
    alt.X('created_year:O', title = 'Created year'),
    alt.Y('channel_type:N', title = 'Channel Type'),
    alt.Color('value_of_chosen_attribute:Q', title = 'Selected Attribute'),
    tooltip =  alt.Tooltip(['value_of_chosen_attribute:Q'])
).transform_calculate(
    value_of_chosen_attribute=f'datum[{color_param.name}]'
).add_params(
    color_param
).properties(title = 'Insight of Earnings, Number of Uploads and subscribers for each Channel Type by Year')
```
